```python
import numpy as np
import astropy.units as u

from plotly.offline import init_notebook_mode

from einsteinpy.plotting import GeodesicPlotter
from einsteinpy.coordinates import SphericalDifferential, CartesianDifferential
from einsteinpy.bodies import Body
from einsteinpy.geodesic import Geodesic
init_notebook_mode(connected=True)
```

```python

sph_obj1 = SphericalDifferential(90*u.m, 0.5*np.pi/2*u.rad, -np.pi/8*u.rad,
                                0*u.m/u.s, 1000*u.rad/u.s, 1900*u.rad/u.s)
Attractor1 = Body(name="BH", mass=6e24 * u.kg, parent=None, R=500*u.m, differential=sph_obj1)

Attractor = Body(name="BH", mass=6e24 * u.kg, parent=Attractor1)
sph_obj = SphericalDifferential(90*u.m, np.pi/2*u.rad, -np.pi/8*u.rad,
                                0*u.m/u.s, 1000*u.rad/u.s, 1900*u.rad/u.s)

Object = Body(differential=sph_obj, parent=Attractor)
geodesic = Geodesic(body=Object, time=0 * u.s, end_lambda=0.002, step_size=5e-8)

obj = GeodesicPlotter()
obj.plot(geodesic)
obj.show()
```
